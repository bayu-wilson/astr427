# Homework 5: Monte Carlo on GPUs

Write a (non-GPU) program to calculate π by integrating the area of a
circle of radius 1 with the Monte-Carlo method. You may use a system provided
random number generator. The Python random module says
it uses the Mersenne Twister which I mentioned in class as a quality
generator. Plot how the error in the estimated value of π changes with
the number of points used to evaluate the integral; does this follow the
expected error for a Monte Carlo method?

```python
import numpy as np, matplotlib.pyplot as plt
```

### Step 1)

Draw randomly selected points from using bounds that creates a shape that is easy to find the area of. In this example I will choose a square.

```python
N = int(1e3)
np.random.seed(123)
draws = np.random.uniform(-1,1,N)
index = np.linspace(-1,1,N)
fig,ax = plt.subplots(1)
ax.scatter(index,draws)
ax.set_aspect('equal')
```

### Step 2)

The ratio of a circle and a square is below:

## $\frac{A_{circle}}{A_{square}} = \frac{\pi r^2} {(2r)^2} = \frac{\pi r^2}{4r^2} = \frac{\pi}{4} $

The area should be directly proportional to the number of points in the circle and square respectively.

```python
N = int(1e3)
np.random.seed(123)
draws = np.random.uniform(-1,1,N)
index = np.linspace(-1,1,N)
mask = index**2 + draws**2 < 1
index1,draws1 = index[mask],draws[mask]
index2,draws2 = index[~mask],draws[~mask]
fig,ax = plt.subplots(1)
ax.scatter(index1,draws1)
ax.scatter(index2,draws2)
ax.set_aspect('equal')
```

### Step 3)

Find ratio and multiply by 4 to find value of pi

```python
len(index1) / N * 4
```
